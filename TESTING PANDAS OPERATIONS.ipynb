```python
Today=pd.to_datetime('2025-02-12')
df['Years_with_Company'] =(Today - df['Join_Date']).dt.days // 365
print(df)
```

```python
df_sorted=df.sort_values(by='Years_with_Company', ascending=False)
```

```python
df_sorted
```

```python
%matplotlib inline
```

```python
df_sorted.plot(y='Salary',x='Years_with_Company',kind='scatter')
```

```python
df_sorted.plot(y='Salary',x='Years_with_Company',kind='bar')
```

```python
df_sorted.plot(y='Salary',x='Years_with_Company',kind='pie')
```
